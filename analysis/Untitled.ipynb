# SMA strategy

ema5 > ema10: long

ema < ema10: close

## tuning parameters

```python
def tuning_parameter(short_window, long_window, mid):
    for col in mid.columns:
        mid["EMA_" + str(short_window) + "_" + col] = mid[col].ewm(
            span=short_window, adjust=False).mean()
        mid["EMA_" + str(long_window) + "_" + col] = mid[col].ewm(
            span=long_window, adjust=False).mean()
    ticker_ls = [col for col in mid.columns if "EMA" not in col]
    for ticker in ticker_ls:
        mid["signal_" + ticker] = np.where(
            mid["EMA_" + str(short_window) + "_" + ticker] >
            mid["EMA_" + str(long_window) + "_" + ticker], 1, -1)
        mid["signal_" + ticker] = mid["signal_" + ticker].shift()
        mid["return_" + ticker] = np.log(mid[ticker] / (mid[ticker].shift()))
    mid.dropna(inplace=True)
    for ticker in ticker_ls:
        mid["strategy_" +
            ticker] = mid["return_" + ticker] * mid["signal_" + ticker]
    strategy_return = mid[[col for col in mid.columns if "strategy" in col]]
    daily_return = ((strategy_return + 1).cumprod(axis=1) - 1).iloc[:, -1]
    return np.mean(daily_return) / np.std(daily_return), np.mean(daily_return)
```

```python
sharp_dict = {}
for short in range(1, 185):
    for long in (short + 1, short+185):
        mid = dow30_20230202.iloc[:, 120:].copy()
        sharp_dict[(short, long)] = tuning_parameter(short, long, mid)
        
```
